```python
import numpy as np
import matplotlib.pyplot as plt
import numpyro as pyro
import numpyro.distributions as dist
from numpyro import handlers
from numpyro.infer import MCMC, NUTS
from numpyro.infer.reparam import TransformReparam
from scipy import stats
from scipy import special as sp
from jax import numpy as jnp
from jax import random
from jax.nn import softmax
import pandas as pd
from jax import numpy as jnp 
import arviz as az
import seaborn as sns ; sns.set()
from numpyro.contrib.control_flow import scan
```

We'll need choice vectors 

$$ 
v_{t+1} = v_{t} + \alpha  (r_{t} - v_t * c_t) \\
c_{t+1} \sim Bernoulli( \sigma(\beta * v_{t+1}) ) 
$$
where values, choices, and rewards have shape arm by subject. Multiplying by choice ensures that unchosen arms will not have their value functions updated 

```python
def draw_priors(n_subjs = 4) : 
    alpha_pr = pyro.sample('alpha_pr', dist.Beta(1, 9))
    alphas = pyro.sample('alpha', dist.Normal(alpha_pr, 0.03).expand((n_subjs,)))
    beta_pr = pyro.sample('beta_pr', dist.Gamma(3,1))
    betas = pyro.sample('beta', dist.Normal(beta_pr, .5).expand((n_subjs,)))
    return (alphas, betas)
```

```python
def simulate(trials = 70, subjs = 10, n_arms = 4, sig = 2, global_loc = 5, global_sig = 2.5) : 
    '''Draw priors for alphas and betas. Simulate n trials '''
    rew_probs_means = stats.norm(loc = global_loc, scale = global_sig).rvs(size = n_arms)
    alphas, betas = handlers.seed(draw_priors, rng_seed=0)(n_subjs = subjs)
    params = {'rew_probs' : rew_probs_means, 'alpha' : alphas, 'beta' : betas}
    data = pd.DataFrame(columns = ['choice', 'reward', 'subject'])
    for s in range(subjs) : 
        values = np.ones(n_arms) * global_loc
        alpha, beta = alphas[s], betas[s]
        subj_dat = pd.DataFrame(columns = ['choice', 'reward', 'subject'])
        for t in range(trials): 
            choice_index = np.random.choice(np.arange(n_arms), p = sp.softmax(beta * values)) 
            choice = np.zeros(n_arms) #choice vector. all zeros save for index where it was picked
            choice[choice_index] = 1
            
            reward = stats.norm(loc = rew_probs_means[choice_index], scale = sig).rvs()
            rew_vec = np.zeros(n_arms)
            rew_vec[choice_index] = reward
            
            values = values + alpha * (rew_vec - values * choice)
            subj_dat = subj_dat.append({'choice' : choice_index,
                                       'reward' : reward, 
                                        'subject' : s}, ignore_index = True)
        data = data.append(subj_dat, ignore_index = True)
    return data, params
df, ps = simulate(trials = 50, subjs = 80)
```

```python
def make_tensors(df) : 
    '''one hot code chocies and generate reward vectors which are zero everywhere except for 
    the index of that trial's choice'''
    one_hot_codes = []
    rew_data = []
    cats = []
    one_hot_df = pd.get_dummies(df.drop('reward', axis = 1),
                                prefix = ['choice'], columns = ['choice'])
    rewards = one_hot_df.copy() # below we scale the one hot df by reward value 
    rewards.loc[:, rewards.columns != 'subject'] = df.reward.values[..., None] * one_hot_df.drop('subject', axis = 1)
    for s in df.subject.unique() : 
        code = one_hot_df[one_hot_df.subject == s].drop('subject', axis = 1)
        rew = rewards[rewards.subject == s].drop('subject', axis = 1)
        one_hot_codes.append(code)
        rew_data.append(rew)
        cats.append(df[df.subject == s].choice)
    return np.stack(one_hot_codes, axis = -1), np.stack(rew_data, axis = -1), np.stack(cats, axis = -1)
```

```python
print(f'params {ps}')
one_hot, rewards, cats = make_tensors(df)
rewards.shape, one_hot.shape
```

the fundamental idea of the model is that, at each successive time step, the agent makes a choice between $n$ arms according to softmax probabilities – this is why we use categorical dist: 
$$
p(c_\psi) = \frac{e^{\beta Q_\psi}}{\sum^n_{i = 1}{e^{\beta Q_i}}} 
$$
After each time step we update values $Q_i$ according to the following rule: 
$$
\bf{Q_{t+1}} = \bf{Q_t} + \alpha (\bf{R_t} - \bf{Q} \odot \bf{C_t})
$$
Where Q, R, C are vectors containing values and one-hot-coded rewards and choices (Hence the make_tensors function), and $\odot$ denotes Hadamard/elementwise product. Thus, the problem is to scan through trials while ensuring correct $R^{(arm \: x \: subject)}$ matrix broadcasting

```python
one_hot.shape
```

```python
def delta(one_hot_codes, rew_data, cat_choices) : 
    n_subjs = one_hot_codes.shape[-1]
    n_arms = one_hot_codes.shape[-2]
    subject_plate = pyro.plate('subjects', n_subjs, dim = -1)
    alpha_mu = pyro.sample('alpha_mu', dist.Uniform(low = 0, high = 1.0))
    alpha_sig = pyro.sample('alpha_sig', dist.HalfCauchy(1))
    beta_mu = pyro.sample('beta_mu', dist.Uniform(low = 0, high = 12.0))
    beta_sig = pyro.sample('beta_sig', dist.HalfCauchy(1))
    with subject_plate : 
        with pyro.handlers.reparam(config={'alpha' : TransformReparam(), 
                                           'beta' : TransformReparam()}) :
            alpha = pyro.sample('alpha', dist.TransformedDistribution(dist.Normal(0,1), 
                                       dist.transforms.AffineTransform(alpha_mu, alpha_sig)))
            beta = pyro.sample('beta', dist.TransformedDistribution(dist.Normal(0,1),
                                     dist.transforms.AffineTransform(beta_mu, beta_sig)))
    
    def transition(carry, outcome) : #values/choices  have shape trial x arms x subjects 
        '''carry over the outcome. Use previous outcome to update values. Thus we initialize to 0'''
        values, prev_choice, prev_reward, = carry
        choices, rewards, cat_choice = outcome
        with subject_plate : 
            values = values + alpha * (prev_reward - values * prev_choice) #make sure params broadcast correctly
            choice_probs = pyro.sample('choice_probs', 
                                       dist.Categorical(
                                       logits = (beta * values).T), #swap axes so that the outer axis is the 'list' of categorical probabilities, which yields a dim = 10 distribution to plate thru
                                       obs = cat_choice) # use categorical codings
            
        return (values, choices, rewards), None 
    
    initial_values = jnp.broadcast_to((jnp.ones((n_arms), dtype = jnp.float32) * 5.)[..., None], 
                                      (n_arms, n_subjs))#.astype('float32')
    scan(transition, (initial_values, jnp.zeros((n_arms, n_subjs), dtype = jnp.float32),#.astype('uint8'), 
                      jnp.zeros((n_arms, n_subjs), dtype = jnp.float32)), 
                     [one_hot_codes, rew_data, cat_choices])

```

```python
data = {'one_hot_codes' : one_hot.astype('float'), 
        'rew_data' : rewards.astype('float'), 'cat_choices' : cats.astype('int')}
rng_key = random.PRNGKey(36)
kernel = NUTS(delta)
mcmc = MCMC(kernel, 
           num_samples = 2000,
           num_warmup = 1000)
mcmc.run(rng_key, **data)
```

```python
mcmc.print_summary()
```

```python
loaded = az.from_numpyro(mcmc)
strong_priors = ['beta_a', 'beta_b', 'alpha_a', 'alpha_b']
weak_priors = ['alpha_mu', 'alpha_sig', 'beta_mu', 'beta_sig']
az.plot_trace(loaded, var_names = weak_priors) ; plt.show()
```
